```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(20)
y = x
plt.figure()
plt.scatter(x,y)
```

```python
colors  = ['green']*(len(x)-1)
colors.append('red')
plt.figure()
plt.scatter(x,y,c=colors,s=100)
```

```python
plt.figure()
plt.scatter(x[:6],y[:6],c='blue',s=100,label='Some x_y blah blah')
plt.scatter(x[6:],y[6:],c='red',s=80,label = 'Some x1_y1 blah blah')
```

```python
#Now let's add axes labels
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('It is a Scatter Plot')
```

```python
plt.legend()
```

```python
plt.legend(loc=4,frameon=False,title='legend') # 4 -  bottom right , frameon - we can manipulate frame
```

# Line Plots

```python
lineardata = np.arange(1,10)
quad_data = lineardata**2
plt.figure()
plt.plot(lineardata,'-o')  # INdex of the series is considered as x axis
```

```python
plt.plot(quad_data,'--r') # Red falg '-'
```

```python
plt.plot(quad_data+1,marker ='o')
```

```python
plt.legend(['lindata','quad_data','quad+1'],loc=2,frameon=False)
```

```python
plt.gca().fill_between(range(len(lineardata)),lineardata,quad_data, # lindata - min , quad -max
                       facecolor = 'black',alpha = 0.4)
```

# Bar plots

```python
plt.figure()
xvals = [_ for _ in range(len(lineardata))]
plt.bar(xvals,lineardata,color = 'blue',width=0.2)
```

```python
x_vals_offset = [_+0.2 for _ in xvals]
plt.bar(x_vals_offset,lineardata,color='red',width=0.2,bottom=[2 for _ in xvals])
```

```python
#Horizontal bar
plt.figure()
plt.barh(xvals,lineardata,color='black',height=0.2) # width and height are different for barh and bar plots
plt.barh(x_vals_offset,lineardata,color='blue',height=0.2)
plt.xlabel('Length')
plt.ylabel('measures')
```

```python
#Exercise
Prog_lang = ['Python','sql','Java','c++','JavaScript']
Popularity = [56,39,34,34,29]
```

```python
positions = np.arange(len(Prog_lang))
plt.figure()
bars = plt.bar(positions,Popularity,align='center',color = list('rgbkymc'),width=0.5,
        alpha= 0.6)
```

```python
plt.xticks(positions,Prog_lang,rotation =20) # setting ticks
```

```python
plt.xlabel('Progamming_lang')
```

```python
# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(bottom=False, top=False, left=False, right=False,labelleft = False,labelcolor = 'green')
```

```python
# remove the frame of the chart.
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
# direct label each bar with Y axis values
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11,rotation=90)
```
