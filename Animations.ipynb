```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

```python
n = 100
y = np.random.normal(loc=0.0,scale=1.0,size=10000)
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins,color='k')
    plt.axis([-4,4,0,30]) # range of axis
    plt.gca().set_title('Sampling the normal distribution')
    plt.gca().set_xlabel('values')
    plt.gca().set_ylabel('Freq')
    plt.annotate('n = {}'.format(curr),[2,20])# we put the text wherever we want
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig,update,interval=100)
```
