# Heat Maps

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.random.normal(loc=0.0,scale=1.0,size=10000)
x = np.random.random(size=10000)
plt.figure()
plt.hist2d(x,y)
```

```python
plt.figure()
plt.hist2d(x,y,bins=100)
```

```python
# add a color bar
plt.colorbar()
```
