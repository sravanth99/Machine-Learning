# Box and Whisker Plots :

```python
import pandas as pd
normal = np.random.normal(loc=0.0,scale = 1.0,size=10000)
gamma = np.random.gamma(2,size=10000)
random = np.random.random(size=10000)
df = pd.DataFrame({'normal':normal,'gamma':gamma,'random':random})
```

```python
df.head()
```

```python
df.describe()
```

```python
# Mean or median of the data is plotted as a straight line
plt.figure()
box = plt.boxplot(df['normal'],whis='range')
```

```python
plt.clf()# cLear the current figure
plt.boxplot([df['random'],df['normal'],df['gamma']],whis='range') # we have to pass the list of columns to it
```

```python
plt.figure()
# we detect the outliers by putting whis parameter off
plt.boxplot([df['random'],df['normal'],df['gamma']])
```
