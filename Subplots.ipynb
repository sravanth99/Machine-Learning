```python
import matplotlib.pyplot as plt
import numpy as np
```

# Subplots: 

```python
#subplot documentation
plt.subplot?
```

```python
plt.figure()
#Let's create a subplot with 1 row and 2 columns  (1,2) and let's use 1st column to plot (1,2,1)
#Last element in the tuple represents cell number
plt.subplot(1,2,1) 
lin_data = np.arange(1,10)
plt.plot(lin_data,'-o')
```

```python
exponential = lin_data**2
sub2 = plt.subplot(1,2,2)
sub2.plot(exponential,'-o')
```

```python
sub2.plot(lin_data,'-o')
```

<p>Both linear plots are same but their areas seem different which often misleads users</p>
<p>To solve this problem let us lock the axes for subplots so that they share the same axes.</p>

```python
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(lin_data,'-o')
ax2 = plt.subplot(1,2,2,sharey = ax1)
plt.plot(exponential,'-o')
```

# 3X3 grid with all axes locked:
<p> We are going to use subplots

```python
fig ,((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9)) = plt.subplots(3,3,sharex=True,sharey=True)
ax1.plot(lin_data,'-o'),ax5.plot(exponential,'-'),ax9.plot(lin_data,':')
```
