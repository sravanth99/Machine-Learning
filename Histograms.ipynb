```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True,sharey=True)
axs = [ax1,ax2,ax3,ax4]
for n in range(0,len(axs)):
    sam_size  = 10**(n+2)
    sample = np.random.normal(loc =0.0,scale= 1.0,size = sam_size)
    axs[n].hist(sample,bins=100)
    axs[n].set_title('n = {}'.format(sam_size))
```

# Bins

```python
size  = 10000
Y = np.random.normal(loc = 0.0,scale =1.0,size=size)
X = np.random.random(size=size)
plt.figure()
plt.scatter(X,Y)
```

```python
#Let's define 3X3 grid spec
#The GridSpec allows you to map axes over multiple cells in a grid.
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3,3)
top_hist = plt.subplot(gspec[0,1:])
side_hist = plt.subplot(gspec[1:,0])
bottom_scatter = plt.subplot(gspec[1:,1:])
```

```python
top_hist.hist(Y,bins=100)
side = side_hist.hist(Y,bins=100,orientation='horizontal') 
# we used orientation parameter to fit the histogram into appropriate shape
bottom_scatter.scatter(X,Y,color = 'r',alpha=0.1)
```
